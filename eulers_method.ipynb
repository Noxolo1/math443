```python
import numpy as np 
import sys
import matplotlib.pyplot as plt

def f(x,t):
    return 1 + x/t

def y(t):
	return t*(1 + np.log(t))

a = 1
b = 6
N = 10
h = (b-a)/N

t = np.zeros(N+1)
t[0] = 1
w = np.zeros(N+1)
w[0] = 1

print('%1.2f %1.5f %1.5f' % (t[0], w[0], abs(w[0] - y(t[0]))))
for j in range(N):
	t[j+1] = t[j] + h
	w[j+1] = w[j] + h*f(t[j], w[j])
	print('%1.2f %1.5f %1.5f' % (t[j+1], w[j+1], abs(w[j+1] - y(t[j+1]))))

plt.figure(figsize = (12, 8))
plt.plot(t, w, 'bo--', label='Approximate')
plt.plot(t, y(t), 'g', label='Exact')
plt.title('Approximate and Exact Solution for Simple ODE')
plt.xlabel('t')
plt.ylabel('f(t)')
plt.grid()
plt.legend(loc='lower right')
plt.show()
```
