```python

import numpy as np
import sys

# Function f that is the subject of the area problem
def f(x):
	return x**3

#####	
# Midpoint approximation
def doMidpoint(a, b, n):
	# identify step size
	h = (b - a)/n

	# initialize sum and step through (use zero to n-1, pesky python)
	sum = 0
	for k in range(0, n):
		xk = a + k*h
		xkp1 = xk + h
		xkm = (xk + xkp1)/2
		
		sum = sum + f(xkm)
	# Finish approximation and return Mh
	Mh = h*sum
	return Mh

#####
# Trapezoidal approximation
def doTrapezoidal(a, b, n):
	# identify step size
	h = (b - a)/n
	
	# initialize sum and step through (use zero to n-1, pesky python)
	sum = 0
	for k in range(0, n):
		xk = a + k*h
		xkp1 = xk + h

		sum = sum + (f(xk) + f(xkp1))
	# Finish approximation and return Th
	Th = h/2*sum
	return Th


def doSimpson(a, b, n):
    h = (b-a)/n
    sum = 0
    for k in range(0,n):
        xk = a + k*h
        xkp1 = xk + h
        xkm = xk + xkp1
        sum = sum + (f(xk) + 4*f((xkm)/2) + f(xkp1))
    s = (h/6)*sum
    return s

yM = []
yT = []
yS = []

for i in range(1, 6):
	xn = doMidpoint(0,1, i**2)
	yM.append(np.abs(np.log(2) - xn))
	xn1 = xn

for i in range(1, 6):
	xn = doTrapezoidal(0,1, i**2)
	yT.append(np.abs(np.log(2) - xn))
	xn1 = xn

for i in range(1, 6):
	xn = doSimpson(0,1, i**2)
	yS.append(np.abs(np.log(2) - xn))
	xn1 = xn

print(yM[len(yM)-1])
print()
print(yT[len(yT)-1])
print()
print(yS[len(yS)-1])

# error arrays
yME = []
yTE = []
ySE = []

# errors 
for i in range(1, len(yM)):

	if((i + 1) >= len(yM)):
		break

	yME.append(yM[i + 1]/ yM[i])
	print()
	yTE.append(yT[i + 1]/ yT[i])
	print()
	ySE.append(yS[i + 1]/ yS[i])

```
